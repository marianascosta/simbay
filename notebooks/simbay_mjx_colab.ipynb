# Simbay MJX on Colab

This notebook clones or updates the repo, installs the minimum packages, and runs the existing `main.py` flow through a small Colab wrapper that forces headless MJX mode.

```python
from pathlib import Path
import os
import subprocess
import sys

REPO_URL = "https://github.com/marianascosta/simbay.git"
REPO_DIR = Path("/content/simbay")

if REPO_DIR.exists():
    subprocess.run(["git", "-C", str(REPO_DIR), "pull"], check=True)
else:
    subprocess.run(["git", "clone", REPO_URL, str(REPO_DIR)], check=True)

os.chdir(REPO_DIR)
print(f"Working directory: {REPO_DIR}")
```

```python
subprocess.run(
    [
        sys.executable,
        "-m",
        "pip",
        "install",
        "-q",
        "mujoco==3.6.0",
        "mujoco-mjx==3.6.0",
        "jax==0.5.0",
        "jaxlib==0.5.0",
        "matplotlib",
        "numpy",
    ],
    check=True,
)
print("Dependencies installed.")
```

```python
os.environ["SIMBAY_HEADLESS"] = "1"
os.environ["SIMBAY_USE_MJX"] = "1"
os.environ["MPLBACKEND"] = "Agg"
os.environ["SIMBAY_PARTICLES"] = "100"

result = subprocess.run(
    [sys.executable, str(REPO_DIR / "colab_main.py")],
    cwd=REPO_DIR,
    text=True,
    capture_output=True,
)
print(result.stdout)
if result.returncode != 0:
    print(result.stderr)
    result.check_returncode()
```
